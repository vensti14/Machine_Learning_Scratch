```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from ISLP import load_data
from sklearn.linear_model import LinearRegression
```

```python
credit = load_data('Credit')
df = credit.copy()
```

```python
X = df[["Income"]].values
y = df[["Balance"]].values

model = LinearRegression()
model.fit(X, y)


print("Intercept (b0) = ", model.intercept_)
print("Slope (b1) = ", model.coef_)

y_pred = model.predict(X)

```

```python
plt.scatter(X, y, c="green", alpha=0.5, label="Data")
plt.plot(X, y_pred, c="black", linewidth=2, label="Sklearn Regression")
plt.xlabel("Income ($1000s)")
plt.ylabel("Balance ($)")
plt.title("Linear Regression with scikitlearn")
plt.legend()
plt.show()
```
