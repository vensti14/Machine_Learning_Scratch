Given customer information from a credit card dataset, predict a customer’s credit card balance using statistical learning methods.

```python
!pip install ISLP
import numpy as np
import pandas as pd
from ISLP import load_data


df = load_data('Credit')

print(df.head())
```

```python
df.info
```

```python
X = df[['Income', 'Limit', 'Rating', 'Cards', 'Age']].values
y = df['Balance']

print(X)
```

```python
np.random.seed(11)
indices = np.random.permutation(X.shape[0])
split = int(0.75 * X.shape[0])
```

```python
train_idx, test_idx = indices[:split], indices[split:]

X_train, X_test = X[train_idx], X[test_idx]
y_train, y_test = y[train_idx], y[test_idx]

X_mean = X_train.mean(axis=0)
X_std  = X_train.std(axis=0)
```

```python
X_train_scaled = (X_train - X_mean) / X_std
X_test_scaled  = (X_test  - X_mean) / X_std
```

```python

X_train_b = np.c_[np.ones(X_train_scaled.shape[0]), X_train_scaled]
X_test_b  = np.c_[np.ones(X_test_scaled.shape[0]),  X_test_scaled]
```

```python
#linear regression

theta = np.linalg.inv(X_train.T @ X_train) @ X_train.T @ y_train
```

```python
y_prediction = X_test @ theta
```

```python
mse = np.mean((y_test - y_prediction) ** 2)
```

```python
print ( f"learned parameters: theta = {theta}")

print(f"MSE: {mse}")
```
